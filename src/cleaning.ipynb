## Modifying column "Organisation"

```python
column = mission_launches_cleaned['Organisation']
print(column)
```

```python
#modifying column Organization: Arm??e de l'Air.
original_string = "Arm??e de l'Air"
cleaned_string = original_string.replace("??", "e")
mission_launches_cleaned['Organisation'] = mission_launches_cleaned['Organisation'].apply(lambda x: x.replace(original_string, cleaned_string))
unique_values = mission_launches_cleaned['Organisation'].unique()
print(unique_values)
```

## Modifying column Rocket_Status:

```python
#modifying column Rocket_Status:
original_status = ("StatusActive", "StatusRetired")
cleaned_status = (original_status[0].replace("Status", ""), original_status[1].replace("Status", ""))
mission_launches_cleaned['Rocket_Status'] = mission_launches_cleaned['Rocket_Status'].replace({original_status[0]: cleaned_status[0], original_status[1]: cleaned_status[1]}, regex=True)
unique_values_status = mission_launches_cleaned['Rocket_Status'].unique()
print(unique_values_status)
```

## Modifying column "Price"

```python
#modifying column Price for "Price in millions"
mission_launches_cleaned.rename(columns={'Price': 'Price in Millions'}, inplace=True)
print(mission_launches_cleaned.columns)
```

```python
#modifying column Organisation for "Price in millions"
mission_launches_cleaned.rename(columns={'Organisation': 'Agency'}, inplace=True)
print(mission_launches_cleaned.columns)
```

```python
# adding cost of Startship Prototype found on internet
mission_launches_cleaned.at[2, 'Price in Millions'] = 20 
```

## Standarizing column "Date"

```python
#modifying "Date" column
column = mission_launches_cleaned['Date']
print(column)
```

```python
#deleting UTC
mission_launches_cleaned['Date'] = mission_launches_cleaned['Date'].str.replace(' UTC', '')
```

```python
#removing exact hour
mission_launches_cleaned['Date'] = mission_launches_cleaned['Date'].str.replace(r'\b\d{2}:\d{2}\b', "", regex=True)
```

```python
#removing days of the week
mission_launches_cleaned['Date'] = mission_launches_cleaned['Date'].str.replace(r'\b(?:Sun|Mon|Tue|Wed|Thu|Fri|Sat)\b', "", regex=True)
```

```python
# new column only with month and year
mission_launches_cleaned['Date'] = pd.to_datetime(mission_launches_cleaned['Date'])
mission_launches_cleaned['Month'] = mission_launches_cleaned['Date'].dt.to_period('M')
```

```python
# new column only with year
mission_launches_cleaned['Date'] = pd.to_datetime(mission_launches_cleaned['Date'])
mission_launches_cleaned['Year'] = mission_launches_cleaned['Date'].dt.to_period('Y')
```

```python
# keeping as a Location only the countries
mission_launches_cleaned['Country'] = mission_launches_cleaned['Location'].str.split(',').str[-1].str.strip()
```

```python
#getting correct countries
country_mapping = {'New Mexico': 'Mexico', 'Pacific Missile Range Facility': 'USA', 'Gran Canaria': 'Spain', 'Shahrud Missile Test Site': 'Iran'}

mission_launches_cleaned['Country'] = mission_launches_cleaned['Country'].replace(country_mapping)
```
