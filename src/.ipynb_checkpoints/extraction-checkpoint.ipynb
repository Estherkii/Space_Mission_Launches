## Getting BUDGET per AGENCY

```python
url2 = "https://en.wikipedia.org/wiki/List_of_government_space_agencies#:~:text=As%20of%202023%2C%2074%20different,agencies%20and%20six%20international%20agencies."
```

```python
res = requests.get(url2)
soup = BeautifulSoup(res.content, "html.parser")
```

```python
# Extract budget column elements
extracting_budget = soup.find_all("td", {"style": "text-align:right;"})
extracting_budget_processed = [element.getText().replace("\n", "").strip() for element in extracting_budget]
print(extracting_budget_processed)
```

```python
#removing total budget from the list "about 62,00"
final_list = ['23,500', '11,700', '7,430', '3,521', '2,388', '2,011', '1,831', '1,750', '1,424', '739', '701', '604', '460', '394', '260', '177', '150', '120', '107', '103', '87', '75', '68', '64', '63', '54', '50', '38', '36', '35', '17', '15', '9', '4', '1', 'About 62,000']

if 'About 62,000' in final_list:
    final_list.remove('About 62,000')

print(final_list)
```
